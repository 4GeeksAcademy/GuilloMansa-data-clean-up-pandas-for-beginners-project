#### Ejercicio 10

¿Los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" son iguales? (★★☆)

> Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

- Muestra ambos promedios de precio por metro cuadrado
- Escribe en una celda markdown una conclusión sobre ellos

````python
# creo una nueva columna y le asigno el valor de la división entre la columnas price y la columna surface.
# luego localizo la fila valdemorillo y galapagar para sacar ambos promedios
# como es un valor de m2 lo redondeo sin decimales

df['pps'] = df['price'] / df['surface']
valdemorillo = df.loc[df['level5'] == 'Valdemorillo']

mean_valdemorillo = valdemorillo['pps'].mean().round(0)

galapagar = df.loc[df['level5'] == 'Galapagar']

mean_galapagar = galapagar['pps'].mean().round(0)

mean_valdemorillo, mean_galapagar
f"Precio medio por m² en Valdemorillo: {mean_valdemorillo:.0f} €", f"Precio medio por m² en Galapagar: {mean_galapagar:.0f} €"

````

**Observación:** de este resultado podemos observar que los precios en Valdemorillo y en Galapagar difieren en el valor de metros cuadrados. El costo del metro cuadrado de superficie en Galapagar es mayor, esto puede ser por varios factores, ya sea por calidad de los materiales construidos, por la ubicación de los inmuebles o por la calidad de vida en sus alrededores.
